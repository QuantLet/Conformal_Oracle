```python
import numpy as np
import matplotlib.pyplot as plt

# ── Cross-model threshold comparison (w=30) ──
models = ["GPT-3.5", "GPT-4", "GPT-4o"]
colors = ["#90CAF9", "#2196F3", "#1976D2"]

# Mean thresholds across 9 assets
q_V_means = [0.002, 0.024, 0.020]
q_V_stds  = [0.003, 0.008, 0.007]
q_E_means = [0.001, 0.021, 0.009]
q_E_stds  = [0.002, 0.010, 0.005]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, means, stds, ylabel, title in [
    (axes[0], q_V_means, q_V_stds, 'Mean threshold', '(a) q̂_V (VaR correction)'),
    (axes[1], q_E_means, q_E_stds, 'Mean threshold', '(b) q̂_E (ES correction)'),
]:
    bars = ax.bar(range(len(models)), means, yerr=stds, capsize=4,
                  color=colors, alpha=0.85, edgecolor='white')
    for bar, val in zip(bars, means):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                f'{val:.4f}', ha='center', va='bottom', fontsize=9)
    ax.set_xticks(range(len(models)))
    ax.set_xticklabels(models)
    ax.set_ylabel(ylabel)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.axhline(0, color='gray', lw=0.5)

plt.tight_layout(w_pad=3)
plt.show()
```
