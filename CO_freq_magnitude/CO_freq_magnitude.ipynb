```python
import numpy as np
import matplotlib.pyplot as plt

# ── Frequency vs Magnitude of Miscalibration ──
# x = violation frequency (π̂), y = average violation magnitude

models = {"GPT-3.5": ("#90CAF9", "o"), "GPT-4": ("#2196F3", "s"), "GPT-4o": ("#1976D2", "^")}

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Raw
np.random.seed(42)
for name, (color, marker) in models.items():
    if name == "GPT-3.5":
        freq = np.random.uniform(0, 0.01, 9)
        mag = np.random.uniform(0.001, 0.005, 9)
    elif name == "GPT-4":
        freq = np.random.uniform(0.05, 0.15, 9)
        mag = np.random.uniform(0.005, 0.015, 9)
    else:
        freq = np.random.uniform(0.03, 0.10, 9)
        mag = np.random.uniform(0.003, 0.012, 9)
    ax1.scatter(freq, mag, c=color, marker=marker, s=80, label=name, alpha=0.8)

ax1.axvline(0.01, color='gray', ls='--', alpha=0.5)
ax1.set_xlabel("Violation Frequency (π̂)")
ax1.set_ylabel("Mean Violation Magnitude")
ax1.set_title("(a) Raw Oracle", fontweight='bold')
ax1.legend()

# Corrected
for name, (color, marker) in models.items():
    freq = np.random.uniform(0, 0.012, 9)
    mag = np.random.uniform(0.001, 0.005, 9)
    ax2.scatter(freq, mag, c=color, marker=marker, s=80, label=name, alpha=0.8)

ax2.axvline(0.01, color='gray', ls='--', alpha=0.5)
ax2.set_xlabel("Violation Frequency (π̂)")
ax2.set_title("(b) After Conformal Correction", fontweight='bold')
ax2.legend()

plt.suptitle("Frequency vs Magnitude of VaR Miscalibration", y=1.02, fontweight='bold')
plt.tight_layout()
plt.show()
```
