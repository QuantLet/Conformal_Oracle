```python
import numpy as np
import matplotlib.pyplot as plt

# ── Dashboard Template ──
# Each dashboard shows 4 panels for one asset:
# (a) Returns + VaR bounds  (b) Violation timeline
# (c) Score distribution    (d) Coverage summary

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Panel (a): Returns + bounds
ax = axes[0, 0]
np.random.seed(42)
T = 170
r = np.random.standard_t(5, T) * 0.015
ax.plot(r, 'k-', alpha=0.5, lw=0.5)
ax.axhline(-0.03, color='#F44336', ls='--', label='Raw VaR')
ax.axhline(-0.055, color='#4CAF50', ls='--', label='Corrected VaR')
ax.set_title("(a) Returns & VaR Bounds", fontweight='bold')
ax.legend(fontsize=8)

# Panel (b): Violations
ax = axes[0, 1]
viol_raw = r < -0.03
viol_corr = r < -0.055
ax.scatter(np.where(viol_raw)[0], r[viol_raw], color='red', s=20, marker='x', label='Raw violations')
ax.scatter(np.where(viol_corr)[0], r[viol_corr], color='green', s=20, marker='o', label='Corrected violations')
ax.set_title("(b) Violation Timeline", fontweight='bold')
ax.legend(fontsize=8)

# Panel (c): Scores
ax = axes[1, 0]
scores = np.random.normal(0, 0.01, 300)
ax.hist(scores, bins=40, color='#2196F3', alpha=0.7, edgecolor='white')
ax.axvline(np.quantile(scores, 0.99), color='red', ls='--', label='q̂_V')
ax.set_title("(c) Nonconformity Scores", fontweight='bold')
ax.legend(fontsize=8)

# Panel (d): Summary
ax = axes[1, 1]
ax.axis('off')
summary = "S&P 500 — GPT-4 (w=30)\n\nRaw π̂ = 0.0862\nCorrected π̂ = 0.0058\nq̂_V = 0.0287\nBasel: GREEN\nZ₂ = -0.20"
ax.text(0.5, 0.5, summary, ha='center', va='center', fontsize=12,
        bbox=dict(boxstyle='round', facecolor='#E8F5E9', alpha=0.8))
ax.set_title("(d) Summary", fontweight='bold')

plt.suptitle("Asset Dashboard: S&P 500", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
