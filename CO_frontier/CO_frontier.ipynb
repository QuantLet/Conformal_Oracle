```python
import numpy as np
import matplotlib.pyplot as plt

# ── Coverage-Efficiency Frontier ──
# Each point: (empirical coverage, mean VaR width)
models_raw = {
    'GPT-3.5 raw': (0.996, 0.018),
    'GPT-4 raw': (0.914, 0.028),
    'GPT-4o raw': (0.943, 0.025),
}
models_corr = {
    'GPT-3.5 corr': (0.996, 0.020),
    'GPT-4 corr': (0.990, 0.044),
    'GPT-4o corr': (0.990, 0.038),
}

fig, ax = plt.subplots(figsize=(10, 6))
for name, (cov, width) in models_raw.items():
    ax.scatter(cov, width, s=100, marker='x', color='#F44336', zorder=5)
    ax.annotate(name, (cov, width), textcoords="offset points", xytext=(5, 5), fontsize=8)

for name, (cov, width) in models_corr.items():
    ax.scatter(cov, width, s=100, marker='o', color='#4CAF50', zorder=5)
    ax.annotate(name, (cov, width), textcoords="offset points", xytext=(5, 5), fontsize=8)

ax.axvline(x=0.99, color='gray', ls='--', alpha=0.5, label='Target 99%')
ax.set_xlabel("Empirical Coverage")
ax.set_ylabel("Mean VaR Width")
ax.set_title("Coverage-Efficiency Frontier", fontweight='bold')
ax.legend(['Raw', 'Corrected', 'Target 99%'])
plt.tight_layout()
plt.show()
```
