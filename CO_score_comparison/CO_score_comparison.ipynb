```python
import numpy as np
import matplotlib.pyplot as plt

# ── Score Comparison ──
# One-sided: s_t^V = q_lo(t) - r(t)  [only penalises left-tail misses]
# Two-sided: s_t^V = max(q_lo(t) - r(t), r(t) - q_hi(t))  [CQR, both tails]

def one_sided_score(q_lo, realized):
    return q_lo - realized

def two_sided_score(q_lo, q_hi, realized):
    return np.maximum(q_lo - realized, realized - q_hi)

# Example
np.random.seed(42)
T = 400
realized = np.random.standard_t(df=5, size=T) * 0.015
q_lo = np.quantile(realized, 0.01) * np.ones(T) + np.random.normal(0, 0.002, T)
q_hi = np.quantile(realized, 0.99) * np.ones(T) + np.random.normal(0, 0.002, T)

s_one = one_sided_score(q_lo, realized)
s_two = two_sided_score(q_lo, q_hi, realized)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.hist(s_one, bins=50, color='#2196F3', alpha=0.7, edgecolor='white')
ax1.axvline(np.quantile(s_one, 0.99), color='red', ls='--', label=f'q̂_V = {np.quantile(s_one, 0.99):.4f}')
ax1.set_title("One-sided score (Definition 3.2)", fontweight='bold')
ax1.legend()

ax2.hist(s_two, bins=50, color='#FF9800', alpha=0.7, edgecolor='white')
ax2.axvline(np.quantile(s_two, 0.99), color='red', ls='--', label=f'q̂_V = {np.quantile(s_two, 0.99):.4f}')
ax2.set_title("Two-sided score (CQR)", fontweight='bold')
ax2.legend()

plt.suptitle("One-sided scores yield tighter thresholds → less capital waste", y=1.02)
plt.tight_layout()
plt.show()
```
