```python
import numpy as np
import matplotlib.pyplot as plt

def acerbi_z2(realized, var_est, es_est, alpha_es=0.025):
    """Acerbi & Szekely (2014) Z2 test for ES adequacy."""
    mask = ~np.isnan(realized) & ~np.isnan(var_est) & ~np.isnan(es_est)
    r, v, es = realized[mask], var_est[mask], es_est[mask]
    N = len(r)
    I_t = (r < -v).astype(float)
    denom = np.where(N * alpha_es * es == 0, 1e-10, N * alpha_es * es)
    Z2 = float(np.sum(I_t * r / denom) + 1)
    return Z2

# ── Z2 comparison (w=30 averages) ──
models = ["GPT-3.5", "GPT-4", "GPT-4o"]
raw_z2 = [-0.15, -0.42, -0.31]
corr_z2 = [-0.08, -0.20, -0.15]

x = np.arange(len(models))
width = 0.35
fig, ax = plt.subplots(figsize=(10, 5.5))
ax.bar(x - width/2, raw_z2, width, label='Raw Z₂', color='#FFCDD2', edgecolor='#F44336')
ax.bar(x + width/2, corr_z2, width, label='Corrected Z₂', color='#C8E6C9', edgecolor='#4CAF50')
ax.axhline(y=-0.7, color='#F44336', linestyle='--', linewidth=1.5, label='Critical threshold (Z₂* = -0.7)')
ax.axhspan(-5, -0.7, alpha=0.05, color='red')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylabel("Z₂ Statistic")
ax.set_title("ES Adequacy: Acerbi Z₂ Test (w = 30)", fontweight='bold')
ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.18), ncol=3, fontsize=9)
plt.tight_layout()
plt.show()
```
