```python
import numpy as np
import matplotlib.pyplot as plt

# ── Coverage comparison: Raw vs Conformal ──
models = ["GPT-3.5", "GPT-4", "GPT-4o"]
raw_coverage = [0.996, 0.914, 0.943]    # 1 - mean(raw_pi_hat) per model
corr_coverage = [0.996, 0.990, 0.990]   # 1 - mean(corr_pi_hat) per model

x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 5.5))
ax.bar(x - width/2, raw_coverage, width, label='Raw Oracle',
       color='#FFCDD2', edgecolor='#F44336', linewidth=1)
ax.bar(x + width/2, corr_coverage, width, label='After Conformal',
       color='#C8E6C9', edgecolor='#4CAF50', linewidth=1)
ax.axhline(y=0.99, color='#F44336', linestyle='--', linewidth=1.5,
           label='Target coverage (99%)')
ax.set_xticks(x)
ax.set_xticklabels(models, fontsize=10)
ax.set_ylabel("Empirical Coverage")
ax.set_ylim(0.88, 1.005)
ax.set_title("VaR 1% Coverage: Raw Oracle vs Conformal Correction (w = 30)",
             fontsize=13, fontweight='bold')
ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.18), ncol=3, fontsize=9)
plt.tight_layout()
plt.subplots_adjust(bottom=0.15)
plt.show()
```
