```python
import numpy as np
import matplotlib.pyplot as plt

# ── Violation Rate Heatmap (after correction) ──
models = ["GPT-3.5", "GPT-4", "GPT-4o"]
assets = ["CRIX", "S&P 500", "SPGTCLTR", "STOXX", "CACT", "GDAXI", "CBU0.L", "FTSE100", "DJCI"]

pi_matrix = np.array([
    [0.004, 0.000, 0.000, 0.000, 0.006, 0.000, 0.006, 0.006, 0.012],
    [0.006, 0.006, 0.006, 0.006, 0.006, 0.006, 0.006, 0.006, 0.012],
    [0.006, 0.006, 0.006, 0.006, 0.006, 0.006, 0.006, 0.006, 0.012],
])

fig, ax = plt.subplots(figsize=(12, 4))
im = ax.imshow(pi_matrix, cmap='RdYlGn_r', aspect='auto', vmin=0, vmax=0.03)
ax.set_xticks(range(len(assets)))
ax.set_xticklabels(assets, rotation=45, ha='right', fontsize=9)
ax.set_yticks(range(len(models)))
ax.set_yticklabels(models)
for i in range(3):
    for j in range(9):
        ax.text(j, i, f"{pi_matrix[i,j]:.3f}", ha='center', va='center', fontsize=8)
plt.colorbar(im, ax=ax, shrink=0.8, label="π̂ (violation rate)")
ax.set_title("Violation Rates After Conformal Correction (target: 0.01)", fontweight='bold')
plt.tight_layout()
plt.show()
```
