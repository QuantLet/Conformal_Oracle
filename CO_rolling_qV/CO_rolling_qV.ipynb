# CO_rolling_qV — Rolling Conformal Threshold on S&P 500

Rolling 250-day conformal threshold q̂_{V,t} for four representative
forecasters on S&P 500, with realised volatility overlay (Figure 1).

**Output:** `fig_rolling_qv.pdf`/`.png`

```python
"""
Rolling 250-day q_hat_{V,t} on S&P 500 (Figure 1).
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

SCRIPT_DIR = Path('.').resolve()
BASE = SCRIPT_DIR.parent
DATA = BASE / 'cfp_ijf_data' / 'paper_outputs' / 'tables'
FIG_DIR = BASE / 'figures'
OUT = SCRIPT_DIR

C_RED = '#E31E24'
C_BLUE = '#0066CC'
C_PURPLE = '#7B2FBE'
C_GRAY = '#666666'

plt.rcParams.update({
    'font.family': 'serif', 'axes.grid': False,
    'savefig.transparent': True, 'savefig.dpi': 300,
    'axes.spines.top': False, 'axes.spines.right': False,
    'font.size': 10,
})

MODELS = ['Chronos-Small', 'TimesFM-2.5', 'Lag-Llama', 'GJR-GARCH']
COLORS = [C_RED, C_BLUE, C_PURPLE, C_GRAY]

rqv = pd.read_csv(DATA / 'rolling_qv_SP500.csv',
                  index_col=0, parse_dates=True)

fig, ax = plt.subplots(figsize=(10, 7))

for m, c in zip(MODELS, COLORS):
    if m in rqv.columns:
        s = rqv[m].dropna()
        ax.plot(s.index, s.values, color=c, lw=1, label=m, alpha=0.8)

ax.axhline(0, color='black', lw=0.5, ls=':')
ax.set_title(r'Rolling 250-day $\hat{q}_{V,t}$ on S&P 500',
             fontsize=12, fontweight='bold')
ax.set_ylabel(r'$\hat{q}_{V,t}$')
ax.set_xlabel('Date')

if 'rvol' in rqv.columns:
    ax2 = ax.twinx()
    rvol = rqv['rvol'].dropna()
    ax2.fill_between(rvol.index, 0, rvol.values, alpha=0.08, color=C_GRAY)
    ax2.set_ylabel('Realised vol.', color=C_GRAY, fontsize=9)
    ax2.tick_params(axis='y', labelcolor=C_GRAY)
    ax2.spines['top'].set_visible(False)

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          ncol=4, fontsize=9, frameon=False)

FIG_DIR.mkdir(exist_ok=True)
for ext in ['pdf', 'png']:
    fig.savefig(OUT / f'fig_rolling_qv.{ext}', bbox_inches='tight')
    fig.savefig(FIG_DIR / f'fig_rolling_qv.{ext}', bbox_inches='tight')

plt.show()
plt.close(fig)
print('Saved: fig_rolling_qv.pdf/.png')
```
