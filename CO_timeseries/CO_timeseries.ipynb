```python
import numpy as np
import matplotlib.pyplot as plt

# ── Time series: Raw vs Corrected VaR/ES ──
print("This quantlet generates time series plots of:")
print("  - Realized log-returns")
print("  - Raw LLM-VaR and LLM-ES bounds")
print("  - Conformally corrected VaR and ES bounds")
print("  - Violation markers (days where return < -VaR)")
print()
print("Key insight: Conformal correction widens the VaR bound by q̂_V,")
print("eliminating most violations while keeping the bound tight.")
print()

# Synthetic demonstration
np.random.seed(42)
T = 150
returns = np.random.standard_t(df=5, size=T) * 0.015
raw_var = np.abs(returns).rolling_mean_placeholder = 0.025  # placeholder
raw_var = np.full(T, 0.025)
q_hat_V = 0.0287
corr_var = raw_var + q_hat_V

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8), sharex=True)
days = np.arange(T)

for ax, var, title, color in [
    (ax1, raw_var, "Raw Oracle", "#2196F3"),
    (ax2, corr_var, "After Conformal Correction", "#4CAF50"),
]:
    ax.plot(days, returns, 'k-', alpha=0.4, lw=0.5, label='Realized')
    ax.plot(days, -var, color=color, lw=1, label=f'−VaR bound')
    viol = returns < -var
    ax.scatter(days[viol], returns[viol], color='red', s=20, marker='x', zorder=5)
    ax.set_title(f"{title} — {viol.sum()} violations / {T} days", fontweight='bold')
    ax.set_ylabel("Log Return")
    ax.legend(fontsize=8)

plt.tight_layout()
plt.show()
```
