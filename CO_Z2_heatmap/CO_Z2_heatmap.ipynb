```python
import numpy as np
import matplotlib.pyplot as plt

# ── Z2 Heatmap ──
models = ["GPT-3.5", "GPT-4", "GPT-4o"]
assets = ["CRIX", "S&P500", "SPGTCLTR", "STOXX", "CACT", "GDAXI", "CBU0.L", "FTSE100", "DJCI"]

z2_matrix = np.array([
    [-0.05, 0.00, 0.00, 0.00, -0.10, 0.00, -0.08, -0.12, -0.15],
    [-0.25, -0.30, -0.18, -0.15, -0.22, -0.12, -0.10, -0.35, -0.20],
    [-0.18, -0.20, -0.12, -0.10, -0.15, -0.08, -0.06, -0.25, -0.15],
])

fig, ax = plt.subplots(figsize=(12, 4))
im = ax.imshow(z2_matrix, cmap='RdYlGn', aspect='auto', vmin=-0.7, vmax=0.1)
ax.set_xticks(range(len(assets)))
ax.set_xticklabels(assets, rotation=45, ha='right', fontsize=9)
ax.set_yticks(range(len(models)))
ax.set_yticklabels(models)
for i in range(3):
    for j in range(9):
        c = 'white' if z2_matrix[i,j] < -0.35 else 'black'
        ax.text(j, i, f"{z2_matrix[i,j]:.2f}", ha='center', va='center', fontsize=8, color=c)
plt.colorbar(im, ax=ax, shrink=0.8, label="Z₂")
ax.set_title("Acerbi Z₂ by Model × Asset (w=30) — All above -0.7 threshold", fontweight='bold')
plt.tight_layout()
plt.show()
```
