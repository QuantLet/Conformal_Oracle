# CFP_Paper_Tables_Figures

Collects all paper outputs.

```python
import shutil
from pathlib import Path

SRC_TAB = Path('../cfp_ijf_data/paper_outputs/tables')
SRC_FIG = Path('../cfp_ijf_data/paper_outputs/figures')
OUT = Path('results')
OUT.mkdir(exist_ok=True)

# Copy tables
for f in sorted(SRC_TAB.glob('*.tex')):
    shutil.copy(f, OUT / f.name)
    print(f'Table: {f.name}')

# Copy figures
for f in sorted(SRC_FIG.glob('*.pdf')):
    shutil.copy(f, OUT / f.name)
    print(f'Figure: {f.name}')

print(f'\nTotal: {len(list(OUT.glob("*.tex")))} tables, {len(list(OUT.glob("*.pdf")))} figures')
```
