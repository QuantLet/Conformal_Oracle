```python
import numpy as np
import matplotlib.pyplot as plt

# ── Sign Diagnostic ──
# q̂_V > 0: model underestimates risk (needs wider VaR)
# q̂_V < 0: model overestimates risk (VaR already conservative)
# q̂_V ≈ 0: model well-calibrated

models = ["GPT-3.5", "GPT-4", "GPT-4o"]
assets = ["CRIX", "S&P 500", "SPGTCLTR", "STOXX", "CACT", "GDAXI", "CBU0.L", "FTSE100", "DJCI"]

q_matrix = np.array([
    [0.004, 0.000, 0.000, 0.000, 0.006, 0.000, 0.000, 0.004, 0.006],
    [0.026, 0.029, 0.028, 0.023, 0.023, 0.020, 0.018, 0.020, 0.026],
    [0.024, 0.022, 0.022, 0.018, 0.020, 0.014, 0.013, 0.016, 0.025],
])

fig, ax = plt.subplots(figsize=(12, 4))
im = ax.imshow(q_matrix, cmap='RdYlGn_r', aspect='auto')
ax.set_xticks(range(len(assets)))
ax.set_xticklabels(assets, rotation=45, ha='right', fontsize=9)
ax.set_yticks(range(len(models)))
ax.set_yticklabels(models)
for i in range(3):
    for j in range(9):
        c = 'white' if q_matrix[i,j] > np.median(q_matrix) else 'black'
        ax.text(j, i, f"{q_matrix[i,j]:.3f}", ha='center', va='center', fontsize=8, color=c)
plt.colorbar(im, ax=ax, shrink=0.8, label="$\hat{q}_V$")
ax.set_title("Sign Diagnostic: Positive = Under-coverage (needs correction)", fontweight='bold')
plt.tight_layout()
plt.show()
```
