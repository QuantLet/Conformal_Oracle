```python
import numpy as np
import matplotlib.pyplot as plt

# ── VaR Width Comparison ──
methods = ['HistSim', 'EWMA', 'GARCH-LPA', 'GAS-t', 'LLM+CP\n(GPT-4o)', 'LLM+CP\n(GPT-4)']
mean_var = [0.0198, 0.0241, 0.0285, 0.0362, 0.0382, 0.0435]
colors = ['#90CAF9', '#90CAF9', '#2196F3', '#2196F3', '#4CAF50', '#4CAF50']
green_pct = [33, 56, 78, 89, 100, 100]

fig, ax = plt.subplots(figsize=(12, 5))
bars = ax.bar(range(len(methods)), mean_var, color=colors, alpha=0.85, edgecolor='white')
for bar, val, g in zip(bars, mean_var, green_pct):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.001,
            f'{val:.4f}\n({g}% Green)', ha='center', va='bottom', fontsize=8)
ax.set_xticks(range(len(methods)))
ax.set_xticklabels(methods, fontsize=9)
ax.set_ylabel("Mean VaR (1%)")
ax.set_title("Price of Distribution-Free Guarantees: VaR Width Comparison", fontweight='bold')
plt.tight_layout()
plt.show()
```
