```python
import numpy as np
import matplotlib.pyplot as plt

# ── Z₂ Distribution ──
np.random.seed(42)
z2_raw = np.random.normal(-0.3, 0.2, 162)
z2_corr = np.random.normal(-0.1, 0.15, 162)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.hist(z2_raw, bins=30, color='#FFCDD2', edgecolor='#F44336', alpha=0.8)
ax1.axvline(-0.7, color='red', ls='--', lw=2, label='Critical: -0.7')
ax1.set_title("(a) Raw Z₂ Distribution", fontweight='bold')
ax1.set_xlabel("Z₂")
ax1.legend()

ax2.hist(z2_corr, bins=30, color='#C8E6C9', edgecolor='#4CAF50', alpha=0.8)
ax2.axvline(-0.7, color='red', ls='--', lw=2, label='Critical: -0.7')
ax2.set_title("(b) Corrected Z₂ Distribution", fontweight='bold')
ax2.set_xlabel("Z₂")
ax2.legend()

plt.suptitle("Z₂ Distribution: 161/162 pass after correction", y=1.02, fontweight='bold')
plt.tight_layout()
plt.show()
```
