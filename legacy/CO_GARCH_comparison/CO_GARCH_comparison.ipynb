```python
import numpy as np
import pandas as pd

# ── GARCH Benchmark Comparison ──
# Results from Table 5 of the paper (w=30, 9 assets)

benchmarks = pd.DataFrame({
    'Method': ['GARCH-LPA', 'GAS-t', 'EWMA', 'HistSim', 'LLM+CP (GPT-4)', 'LLM+CP (GPT-4o)'],
    'Mean VaR': [0.0285, 0.0362, 0.0241, 0.0198, 0.0435, 0.0382],
    'Green (of 9)': [7, 8, 5, 3, 9, 9],
    'Mean pi_hat': [0.0120, 0.0085, 0.0180, 0.0250, 0.0058, 0.0058],
    'Z2 Pass (of 9)': [8, 9, 6, 4, 9, 9],
})

print("=" * 75)
print("Table 5: Cross-Method VaR Comparison (w=30)")
print("=" * 75)
print(benchmarks.to_string(index=False))
print()
print("Key findings:")
print("  • LLM+CP achieves 100% Green zone (9/9 assets)")
print("  • GARCH ratio: GPT-4 = 1.53× GARCH-LPA, 1.20× GAS-t")
print("  • Price of distribution-free guarantee: 20-53% wider VaR")
print("  • But: NO parametric assumptions required")
```
