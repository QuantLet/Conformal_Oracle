```python
import pandas as pd

# ── Inference Costs ──
costs = pd.DataFrame({
    'Model': ['GPT-3.5 Turbo', 'GPT-4', 'GPT-4o'],
    'Input ($/1M tok)': [0.50, 30.00, 2.50],
    'Output ($/1M tok)': [1.50, 60.00, 10.00],
    'Est. cost/asset/day': ['$0.02', '$1.20', '$0.10'],
    'Time/call (s)': ['~3', '~15', '~5'],
    'Quality': ['Low', 'High', 'High'],
})

print("LLM Inference Costs (1024 samples, ~150 token context)")
print("=" * 75)
print(costs.to_string(index=False))
print()
print("9 assets × 250 days/year:")
print("  GPT-3.5: ~$45/year   (but poor calibration)")
print("  GPT-4:   ~$2,700/year (best quality)")
print("  GPT-4o:  ~$225/year  (best cost-quality ratio)")
print()
print("Conformal correction adds negligible cost (pure NumPy, <1s)")
```
