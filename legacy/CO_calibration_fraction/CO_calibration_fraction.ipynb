```python
import numpy as np
import matplotlib.pyplot as plt

# ── Calibration Fraction Sensitivity ──
f_cals = [0.50, 0.55, 0.60, 0.65, 0.70, 0.75, 0.80, 0.85, 0.90]
# Simulated results for GPT-4, S&P 500
coverages = [0.985, 0.987, 0.988, 0.989, 0.990, 0.991, 0.992, 0.993, 0.995]
var_widths = [0.038, 0.040, 0.041, 0.042, 0.044, 0.046, 0.049, 0.053, 0.060]

fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.plot(f_cals, coverages, 'o-', color='#4CAF50', lw=2, label='Coverage')
ax1.axhline(0.99, color='gray', ls='--', alpha=0.5)
ax1.set_xlabel("Calibration Fraction (f_cal)")
ax1.set_ylabel("Empirical Coverage", color='#4CAF50')
ax1.set_ylim(0.98, 1.0)

ax2 = ax1.twinx()
ax2.plot(f_cals, var_widths, 's-', color='#2196F3', lw=2, label='VaR Width')
ax2.set_ylabel("Mean VaR Width", color='#2196F3')

ax1.axvline(0.70, color='#F44336', ls='--', alpha=0.5, label='Chosen: f_cal = 0.70')
ax1.set_title("Calibration Fraction Sensitivity (GPT-4, S&P 500)", fontweight='bold')
fig.legend(loc='upper center', bbox_to_anchor=(0.5, -0.02), ncol=3)
plt.tight_layout()
plt.show()

print("f_cal = 0.70 balances: enough calibration data + enough test data for evaluation")
```
